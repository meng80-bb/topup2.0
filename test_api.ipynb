```python
import requests

base_url = "http://127.0.0.1:5000/api/tasks"
params = {
    "user_id": "user123",
    "limit": 10,
    "offset": 0,
}

resp = requests.get(base_url, params=params)
print(resp.status_code, resp.text)
```

```python
import requests

url = "http://127.0.0.1:5000/api/tasks"  # 修改为实际的地址
payload = {
    "user_id": "user123",
    "workflow_name": "get_available_dates_workflow",
    "task_name": "获取可用日期列表",
    "parameters": {
        "include_processed": False,
        "user_id": "user123",
        "sort_order": "asc",
        "round": "round18"
    }
}

resp = requests.post(url, json=payload)  # requests 会自动设置 Content-Type: application/json
print(resp.status_code, resp.text)
```

```python
import requests
import json

base_url = "http://127.0.0.1:5000/api/tasks/3"


resp = requests.get(base_url)
# 吧resp.text中的\uXXXX转成正常的中文
resp_text = json.loads(resp.text)
print(resp.status_code, json.dumps(resp_text, ensure_ascii=False, indent=4))
```
